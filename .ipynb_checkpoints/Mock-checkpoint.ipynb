Problem Statement:
Download the Iris flower dataset or any other dataset into a DataFrame. (e.g., https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:

List down the features and their types (e.g., numeric, nominal) available in the dataset.
Create a histogram for each feature in the dataset to illustrate the feature distributions.
Create a box plot for each feature in the dataset.
Compare distributions and identify outliers.

```python
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.columns = iris.feature_names
```

```python
df['class'] = iris.target
```

```python
df['class'] = df['class'].map({0:iris.target_names[0],1:iris.target_names[1],2:iris.target_names[2]})
```

Feature Information:

<br>sepal length in cm type float (numeric)

<br>class type object (nominal):
<br>Iris Setosa
<br>Iris Versicolour
<br>Iris Virginica

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of sepal length")
sns.histplot(df['sepal length (cm)'],color='Blue')
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of sepal width")
sns.histplot(df['sepal width (cm)'],color='Red')
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of petal length")
sns.histplot(df['petal length (cm)'],color='Green')
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of petal width")
sns.histplot(df['petal width (cm)'],color='Yellow')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,12))
fig.suptitle('Distributions of features', fontsize=16)
axs[0][0].set_title('Distribution of sepal length')
sns.kdeplot(x='sepal length (cm)', data=df, ax=axs[0][0], hue='class')
axs[0][1].set_title('Distribution of sepal width')
sns.kdeplot(x='sepal width (cm)', data=df, ax=axs[0][1], hue='class')
axs[1][0].set_title('Distribution of petal length')
sns.kdeplot(x='petal length (cm)', data=df, ax=axs[1][0], hue='class')
axs[1][1].set_title('Distribution of petal width')
sns.kdeplot(x='petal width (cm)', data=df, ax=axs[1][1], hue='class')
```

```python
plt.figure(figsize=(12,8))
plt.title('Box plots of all features in the dataset.')
sns.boxplot(data=df, hue='class')
```

```python
iris_long = df.melt(id_vars=['class'])
ax = sns.boxplot(data=iris_long, x="value", y="variable", orient="h", hue="class")
plt.figure(figsize=(12,8))
plt.show()
```
