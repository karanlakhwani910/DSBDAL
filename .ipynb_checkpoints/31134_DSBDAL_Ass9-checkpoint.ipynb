Data Visualization II
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for
distribution of age with respect to each gender along with the information about whether
they survived or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
sns.boxplot(x = 'Sex', y = 'Age', hue = 'Survived', data = df)
```

#### Men of lower age group survived
#### Woment of higher age group survived

```python
sns.boxplot(x = 'Sex', y = 'Age', hue = 'Pclass', data = df)
```

#### Maximum higher age group passengers belonged to 1st Class

```python
sns.boxplot(x = 'Sex', y = 'Age', hue = 'Embarked', data = df)
```

#### Not much age wise variation for emabrking but maximum males baorded at Q and maximum females at C

```python
sns.boxplot(x = 'Sex', y = 'Fare', hue = 'Survived', data = df)
```

#### Maximum surving passengers paid more

```python
sns.boxplot(x = 'Sex', y = 'Fare', hue = 'Pclass', data = df)
```

#### 1st class had maximum fare and maximum passegers and the reverse of 3rd Class

```python
sns.boxplot(x = 'Sex', y = 'Fare', hue = 'Embarked', data = df)
```

#### Passengers baording at C paid the maximum fare

```python
sns.jointplot(x = "Fare", y = "Age",
              kind = "scatter", data = df)
```
