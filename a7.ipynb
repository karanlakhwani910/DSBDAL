```python
from nltk.tokenize import word_tokenize,sent_tokenize,RegexpTokenizer
```

```python
#sentence tokenisation
tokenized_text=sent_tokenize(text)
print("line 1 : \n",tokenized_text[0])
print("line 2 : \n",tokenized_text[1])
print("line 3 : \n",tokenized_text[2])
```

```python
#word tokenisation
tokenized_words=word_tokenize(text)
print(tokenized_words[:20])
```

```python
#A RegexpTokenizer splits a string into substrings using a regular expression.
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text)

tokens[:20]
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokens)


# Frequency Distribution 
import matplotlib.pyplot as plt
fdist.plot(20,cumulative=False)
plt.show()
```

# Stop Words Removal

Removing noise from text

```python
dir(nltk.corpus)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
tokens_no_stop_words = []

for w in tokens:
    if w not in stop_words:
        tokens_no_stop_words.append(w)

tokens_no_stop_words[:20]
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
stemmed_words = []

for w in tokens_no_stop_words:
    stemmed_words.append(stemmer.stem(w))
    print(w, " -> ",stemmer.stem(w))
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
lemmatized_words = []

for w in tokens_no_stop_words:
    lemmatized_words.append(lem.lemmatize(w))
    print(w, " -> " , lem.lemmatize(w))
```

```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
first_sentence = "Data Science is the coolest job of the 21st century"
second_sentence = "machine learning is the key for data science"
```

```python
text = [first_sentence,second_sentence]
```

```python
vectorizer=TfidfVectorizer()
tfIdf = vectorizer.fit_transform(text)
```

```python
df = pd.DataFrame(np.divide(tfIdf[0].T.todense(),vectorizer.idf_.reshape(-1,1)), index=vectorizer.get_feature_names(), columns=["DF"])
print (df)
```

```python
idf = pd.DataFrame(np.array(vectorizer.idf_), index=vectorizer.get_feature_names(), columns=["IDF"])
print (idf)
```

```python
tfidf = pd.DataFrame(tfIdf[0].T.todense(), index=vectorizer.get_feature_names(), columns=["TF-IDF"])
print (tfidf)
```

```python
final = pd.concat([df,idf,tfidf],axis=1)
```
