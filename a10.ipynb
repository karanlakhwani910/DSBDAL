```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.info()
```

Feature Information:
   1. sepal length in cm type float (numeric)
   2. sepal width in cm type float (numeric)
   3. petal length in cm type float (numeric)
   4. petal width in cm type float (numeric)
   5. class type object (nominal): 
      * Iris Setosa
      * Iris Versicolour
      * Iris Virginica

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of sepal length")
sns.histplot(df['sepal_length'],color='Blue',stat='density')
sns.kdeplot(df['sepal_length'], shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of petal length")
sns.histplot(df['petal_length'],color='Blue',stat='density')
sns.kdeplot(df['petal_length'], shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of sepal_width")
sns.histplot(df['sepal_width'],color='Blue',stat='density')
sns.kdeplot(df['sepal_width'], shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of petal_width")
sns.histplot(df['petal_width'],color='Blue',stat='density')
sns.kdeplot(df['petal_width'], shade=True)
```

```python
df['species'].value_counts()
```

```python
#converting object column to numeric using python dictionary


species_map={'virginica':0,'versicolor':1,'setosa':2}
df['species']=df['species'].map(species_map)
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of species")
sns.histplot(df['species'],color='Blue',stat='density')
sns.kdeplot(df['species'], shade=True)
```

```python
iris = pd.read_csv('iris.data')
iris_copy = iris.copy()
iris_copy.head()
```

```python
help(plt.subplots)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,12))
fig.suptitle('Distributions of features', fontsize=16)
axs[0][0].set_title('Distribution of sepal length')
sns.histplot(x='sepal length (cm)', data=iris_copy, ax=axs[0][0], kde = True, hue='class')
axs[0][1].set_title('Distribution of sepal width')
sns.histplot(x='sepal width (cm)', data=iris_copy, ax=axs[0][1], kde = True, hue='class')
axs[1][0].set_title('Distribution of petal length')
sns.histplot(x='petal length (cm)', data=iris_copy, ax=axs[1][0], kde = True, hue='class')
axs[1][1].set_title('Distribution of petal width')
sns.histplot(x='petal width (cm)', data=iris_copy, ax=axs[1][1], kde = True, hue='class')
```

```python
plt.figure(figsize=(12,8))
plt.title('Box plots of all features in the dataset.')
sns.boxplot(data=iris_copy)
```
